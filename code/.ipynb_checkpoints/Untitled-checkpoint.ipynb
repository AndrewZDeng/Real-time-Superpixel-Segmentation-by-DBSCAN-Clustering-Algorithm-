```python
%matplotlib widget
import ctypes
import numpy as np
from numpy.ctypeslib import ndpointer
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
lib = ctypes.cdll.LoadLibrary('./DBscan_supel.so')
DBSCAN_supel = lib.DBSCAN_supel
```

```python
DBSCAN_supel.restype = ctypes.c_int
DBSCAN_supel.argtypes = [ndpointer(ctypes.c_int),
                        ctypes.c_int,
                        ctypes.c_int,
                        ndpointer(ctypes.c_int),
                        ctypes.c_int,
                        ctypes.c_int,]
```

```python
image = mpimg.imread('../imgs/frame_11.png')[:,:,0:3] * 255
image = image.astype(np.int32)
```

```python
height = image.shape[0]
width = image.shape[1]
num_supels = 1000
post = 1
labels = np.empty((height,width), dtype=np.int32)
```

```python
real_num_supels = DBSCAN_supel(image, height, width, labels, num_supels, post)
```

```python
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

ax.imshow(mark_boundaries(img_as_float(image/255), labels))
ax.set_axis_off()
```

```python
from skimage.segmentation import slic
```

```python
labels_slic = slic(image, n_segments=5000, compactness=10, sigma=1, start_label=1)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

ax.imshow(mark_boundaries(img_as_float(image/255), labels_slic))
ax.set_axis_off()
```
